```python
from graphviz import Digraph

# Create a directed graph
dot = Digraph()

# Define the nodes
dot.node("A", "Start")
dot.node("B", "Send API Request")
dot.node("C", "Receive JSON Data")
dot.node("D", "Parse JSON Response")
dot.node("E", "Extract Relevant Fields")
dot.node("F", "Store in CSV/Database")
dot.node("G", "Perform Data Analysis / Visualization")
dot.node("H", "End")

# Define the edges (flow)
dot.edge("A", "B")
dot.edge("B", "C")
dot.edge("C", "D")
dot.edge("D", "E")
dot.edge("E", "F")
dot.edge("F", "G")
dot
```
